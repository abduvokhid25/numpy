```python
import numpy as np
arr = np.array([1,3,4,5])
print("numpy array:", arr)
arr.shape
```

```python
arr2 = np.array([[1,3,4,5], [4,1,3,2]])
print("numpy array2:", arr)

arr2.shape
```

```python
#dtype
a = np.array([1,4,6,3])
print (a.dtype)
```

```python
# reshape
a = np.array([1,3,4,5,2,4])
b = a.reshape(2, 3)
print('shaped array:', b)
```

```python
#astype

a = np.array([3,5,3,5,3])
b = a.astype(float)
print('float:', b)
```

```python
# np.zero(), np.one()  => dummy data

z = np.zeros((2,3))
o = np.ones((2, 3))

print('zeros:', z)
print('ones:', 0)
```

```python
#ndim

a = np.array([[2,3,5], [4,2,3]])
print('ndim:', a.ndim)
```

```python
# arrange()
a = np.arange(1, 6)
print("arange:", a)
```

```python
# sum(), mean(), max(), min()

a = np.array([1, 2, 3, 4])

print("sum:", a.sum())
print("mean:", a.mean())
print("max:", a.max())
print("min:", a.min())
```

```python
# flatten()
a = np.array([[1, 2], [3, 4]])
print('flatten:', a.flatten())
```

```python
# T (Transpose)
a = np.array([[1, 2], [3, 4]])
print("transpose:", a.T)
```

```python
# dot() @
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])
print("dot/multiplication:", a.dot(b))
```

```python
# random.rand()

a = np.random.rand(2, 5)
print('random.rand:', a)
```
